Use a write token.

```python
model_path = "/content/MSAE/4096_512_TopKReLU_64_RW_False_False_0.0_KAGL_Fashion-CLIP_train_image_39990_512.pth"
save_dir = "msae-clip-fashion-4096-topk64"

os.makedirs(save_dir, exist_ok=True)

# Copy model weights
shutil.copy(model_path, f"{save_dir}/MSAE_100.bin")
```

```python
config = {
    "architecture": "MSAE",
    "activation": "TopKReLU",
    "k": 64,
    "input_dim": 512,
    "hidden_dim": 4096,
    "dataset": "KAGL Fashion-CLIP",
    "train_images": 39990,
    "modality": "image",
    "sparsity_type": "Matryoshka",
    "reweighted": True
}

with open(f"{save_dir}/config.json", "w") as f:
    json.dump(config, f, indent=2)
```

```python
readme = f"""
# MSAE CLIP Fashion (4096, TopK=64)

This is a **Matryoshka Sparse Autoencoder (MSAE)** trained on **CLIP image embeddings**
from the **KAGL** dataset.

## Model details
- Input dim: 512 (CLIP ViT-B/32 image embeddings)
- Hidden dim: 4096
- Sparsity: Matryoshka
- Dataset size: 39,990 images
- Training objective: reconstruction + sparsity

## Usage
This model is intended for interpretability and feature analysis of CLIP embeddings.

## Training
Trained for 30 epochs.
"""
with open(f"{save_dir}/README.md", "w") as f:
    f.write(readme)
```

```python
from huggingface_hub import HfApi

repo_id = "Alloryne/msae-clip-fashion-4096-topk64"

api = HfApi()
api.create_repo(repo_id, exist_ok=True)

api.upload_folder(
    folder_path=save_dir,
    repo_id=repo_id
)
```
